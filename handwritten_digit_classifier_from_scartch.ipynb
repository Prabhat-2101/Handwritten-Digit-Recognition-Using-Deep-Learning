```python
import cv2
import pandas as pd
import numpy as  np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from PIL import Image
from tensorflow.keras import layers,models
from tensorflow.keras.utils import to_categorical,plot_model
from sklearn.model_selection import train_test_split
```

# Create an custom ResNet50 model.

```python
def resnet50_model(input_shape=(28, 28, 1), num_classes=10):
    input_tensor = layers.Input(shape=input_shape)

    x = layers.Conv2D(64, (7, 7), strides=(2, 2), padding='same')(input_tensor)
    print("BNL",x)
    x = layers.BatchNormalization()(x)
    print("ABNL",x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = _residual_block(x, filters=[64, 64, 256], block_id=1, strides=(1, 1))
    x = _residual_block(x, filters=[128, 128, 512], block_id=2)
    x = _residual_block(x, filters=[256, 256, 1024], block_id=3)
    x = _residual_block(x, filters=[512, 512, 2048], block_id=4)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs=input_tensor, outputs=x, name='resnet50')
    return model
```

```python
def _residual_block(x, filters, block_id, strides=(2, 2)):
    shortcut = x

    x = layers.Conv2D(filters[0], (1, 1), strides=strides, padding='valid', name=f'res_block_{block_id}_conv1')(x)
    x = layers.BatchNormalization(name=f'res_block_{block_id}_bn1')(x)
    x = layers.Activation('relu')(x)


    x = layers.Conv2D(filters[1], (3, 3), padding='same', name=f'res_block_{block_id}_conv2')(x)
    x = layers.BatchNormalization(name=f'res_block_{block_id}_bn2')(x)
    x = layers.Activation('relu')(x)


    x = layers.Conv2D(filters[2], (1, 1), name=f'res_block_{block_id}_conv3')(x)
    x = layers.BatchNormalization(name=f'res_block_{block_id}_bn3')(x)


    if strides != (1, 1) or shortcut.shape[-1] != filters[2]:
        shortcut = layers.Conv2D(filters[2], (1, 1), strides=strides, name=f'res_block_{block_id}_shortcut')(shortcut)
        shortcut = layers.BatchNormalization(name=f'res_block_{block_id}_shortcut_bn')(shortcut)

    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)

    return x
```

# Load the predefined MNIST dataset using Keras

```python
(X_train,Y_train),(X_test,Y_test) = keras.datasets.mnist.load_data()
X_train.shape, X_test.shape
```

```python
X_train[0]
```

```python
X_train = np.expand_dims(X_train,-1)
X_test = np.expand_dims(X_test,-1)
X_train.shape, X_test.shape
```

```python
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
X_train[0]
```

```python
Y_train = to_categorical(Y_train,10)
Y_test = to_categorical(Y_test,10)
Y_train[0]
```

```python
X_train,X_val,Y_train,Y_val = train_test_split(
    X_train,Y_train,test_size = 0.15,random_state=42
)
X_train.shape,X_val.shape
```

```python
model = resnet50_model()
model.summary()
```

```python
plot_model(model,
    to_file='model.png',
    show_layer_names=True,
    show_shapes=True,
    show_dtype=True
)
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
histroy = model.fit(
    X_train,
    Y_train,
    validation_data=(X_val,Y_val),
    epochs=8,
    batch_size=64
)
```

```python
test_loss,test_accuracy = model.evaluate(X_test,Y_test)
print("Test Loss:",test_loss)
print("Test Accuracy:",test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
Y_pred = model.predict(X_test)
Y_pred = np.argmax(Y_pred,axis=1)
Y_true = np.argmax(Y_test,axis=1)

confusion_matrix = confusion_matrix(Y_true,Y_pred)
print(confusion_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix,annot=True,fmt='d',cmap='Blues')
plt.xlabel('Predicted Classes')
plt.ylabel('True Classes')
plt.title('Confusion Matrix')
plt.show()
```
